```python
import numpy as np
import pandas as pd
import os
import glob
import uuid
import base64
from tqdm.auto import tqdm
```

```python
def genIDs(num):
    id_lst = []
    for i in tqdm(range(num)):
        id_lst.append(i)
    return id_lst
```

```python
data_root_dpath = os.path.join("..", "data", "OPP-115")

processed_data_dpath = os.path.join(data_root_dpath, "processed_data")

op_annotations_dpath = os.path.join(processed_data_dpath, "processed_annotations")

master_annotations_115_fpath = os.path.join(processed_data_dpath, "master_annotations_115.csv")

catsplit_annotations_115_parsed_dpath = os.path.join(processed_data_dpath, "catsplit_annotations_115_parsed")

site_metadata_fpath = os.path.join(processed_data_dpath, "site_metadata_115.csv")
```

# Table 1 and 2: Site metadata and site-sector map

```python
site_metadata_df = pd.read_csv(site_metadata_fpath)
site_metadata_df
```

```python
lst = []
for i,r in site_metadata_df.iterrows():
    for j in eval(r['sectors']):
        lst.append(j)
```

```python
unique_sec = list(set(lst))
unique_sec
```

```python
site_metadata_df_orig = site_metadata_df.copy()
```

```python
for sec in unique_sec:
    site_metadata_df[sec] = 0
    
site_metadata_df
```

```python
for i,r in site_metadata_df.iterrows():
    for sec in eval(r['sectors']):
        site_metadata_df.loc[i, sec] = 1
```

```python
site_metadata_df.drop(['sectors'], axis = 1, inplace = True)
site_metadata_df
```

```python
site_sector_map_df = pd.melt(site_metadata_df, id_vars=['policy_ID'], value_vars= site_metadata_df.columns[4:],
        var_name='sector', value_name='sector_val')
site_sector_map_df = site_sector_map_df[site_sector_map_df.sector_val == 1]
site_sector_map_df = site_sector_map_df.drop(["sector_val"], axis = 1).reset_index(drop = True)
site_sector_map_df
```

```python
ids = genIDs(site_sector_map_df.shape[0])
site_sector_map_df['site_sector_map_ID'] = ids
site_sector_map_df = site_sector_map_df[["site_sector_map_ID", "policy_ID", "sector"]]
site_sector_map_df
```

```python
site_metadata_df_orig.drop(["sectors"], inplace = True, axis = 1)
site_metadata_df_orig
```

# Table 3: Category metadata

```python
master_df = pd.read_csv(master_annotations_115_fpath)
master_df
```

```python
lst_cat = list(set(master_df.category))
lst_cat
```

```python
ids = genIDs(len(lst_cat))
```

```python
ids
```

```python
cat_metadata_df = pd.DataFrame(list(zip(ids, lst_cat)), columns = ["category_ID", "category"])
cat_metadata_df
```

# Table 4: Annotation data

```python
master_df = pd.read_csv(master_annotations_115_fpath)
master_df
```

```python
master_df = master_df.merge(cat_metadata_df, how = "inner", on = "category")
master_df
```

```python
master_df.drop(["category", "attr_val"], axis = 1, inplace = True)
master_df
```

# Table 5 and 6: Attribute metadata and attribute category map

```python
cat_attr_lst = []
for fname in glob.glob(r"{}/*.csv".format(catsplit_annotations_115_parsed_dpath)):   
    if not (os.path.basename(fname) in ["Introductory-Generic.csv", "Practice_not_covered.csv", "Privacy_contact_information.csv"]):
        df = pd.read_csv(fname)
        cat_attr_lst.append({col:next(iter(set(df.category))) for col in list(df.columns[6:])})
```

```python
cat_attr_lst
```

```python
attr_cat_map_df = pd.DataFrame(
    [{"attr": key, "category": value} for d in cat_attr_lst for key, value in d.items() ])
ids = genIDs(attr_cat_map_df.shape[0])
attr_cat_map_df['attr_cat_map_ID'] = ids
attr_cat_map_df = attr_cat_map_df[["attr_cat_map_ID", "category", "attr"]]
attr_cat_map_df
```

```python
attr_table_df = attr_cat_map_df[["attr"]].drop_duplicates(ignore_index=True)
attr_table_df
ids = genIDs(attr_table_df.shape[0])
attr_table_df['attr_ID'] = ids
attr_table_df = attr_table_df[["attr_ID", "attr"]]
attr_table_df
```

```python
attr_cat_map_df = attr_cat_map_df.merge(attr_table_df, how = "inner", on = "attr").merge(cat_metadata_df, how = "inner", on = "category")
attr_cat_map_df = attr_cat_map_df[["attr_cat_map_ID", "category_ID", "attr_ID"]]
```

```python
attr_cat_map_df
```

# Table 7: Attribute data

```python
attr_data_df = pd.DataFrame()
for fname in glob.glob(r"{}/*.csv".format(catsplit_annotations_115_parsed_dpath)):   
    df = pd.read_csv(fname)
    attr_data_df = pd.concat([attr_data_df, pd.melt(df, id_vars=['annotation_ID'], value_vars= df.columns[6:],
        var_name='attr', value_name='attr_val')], axis =0, ignore_index = True)
```

```python
attr_data_df
```

```python
attr_data_df = attr_data_df.merge(attr_table_df, how = "inner", on = "attr")
attr_data_df
```

```python
ids = genIDs(attr_data_df.shape[0])
attr_data_df['attr_data_ID'] = ids
attr_data_df = attr_data_df[["attr_data_ID", "annotation_ID", "attr_ID", "attr_val"]]
attr_data_df
```

```python
attr_data_df[["attr_ID", "attr_val"]].drop_duplicates(ignore_index = True).to_csv("test.csv", index = False)
```

```python
# Save 7 tables
op_rdb_dpath = os.path.join(processed_data_dpath, "csv_relational_data")
os.makedirs(op_rdb_dpath, exist_ok = True)
site_metadata_df_orig.to_csv(os.path.join(op_rdb_dpath, "site_metadata.csv"), index = False)
site_sector_map_df.to_csv(os.path.join(op_rdb_dpath, "site_sector_map.csv"), index = False)
cat_metadata_df.to_csv(os.path.join(op_rdb_dpath, "category_metadata.csv"), index = False)
master_df.to_csv(os.path.join(op_rdb_dpath, "annotation_data.csv"), index = False)
attr_table_df.to_csv(os.path.join(op_rdb_dpath, "attr_metadata.csv"), index = False)
attr_cat_map_df.to_csv(os.path.join(op_rdb_dpath, "attr_category_map.csv"), index = False)
attr_data_df.to_csv(os.path.join(op_rdb_dpath, "attr_values_map.csv"), index = False)
```
